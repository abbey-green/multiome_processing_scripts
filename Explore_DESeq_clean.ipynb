```python
# Load files and view cell counts of psuedobulk files 
file_path = 'pb_files_2/'
pb_files = glob.glob(os.path.join(file_path, '*metadata_pfc_022224.csv'))
df_all = pd.DataFrame()

for file in pb_files:
    # print(file)
    celltype_1 = (file.split('_'))[2]
    celltype = (celltype_1.split('/'))[1]
    # print(celltype)
    meta = pd.read_csv(file)
    # display(meta['Unnamed: 0'])
    df_temp = pd.DataFrame()
    df_temp['donor'] = meta['Unnamed: 0']
    df_temp['cell_count'] = meta['cell_count']
    df_temp['celltype'] = celltype
    df_all = pd.concat([df_all, df_temp], ignore_index=True)
    # break

df_all
```
